```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#load the dataset
df = pd.read_csv('PB_All_2000_2021.csv', sep=';')
df
```

```python
print(df.shape)
```

```python
df.info()
```

```python
#Statistcs of the data
df.describe().T
```

```python
#missing values
df.isnull().sum()
```

```python
#change the date format
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df
```

```python
df.info()
```

```python
df = df.sort_values(by=['id', 'date'])
df.head(10)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.head(10)
```

```python
df.columns
```

```python
pollutants = [ 'O2', 'NO3', 'NO2', 'SO4','PO4', 'CL']
```

```python
#WEEK 2
```

```python
#droping missing values
df = df.dropna(subset = pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
#Feature Selection
X = df[['id', 'year']]
Y = df[pollutants]
```

```python
#Encoding
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)
```

```python
#train the model
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, Y_train)
```

```python
#evaluate the model
Y_pred = model.predict(X_test)
```

```python
print("Model Performance on the test data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print(' MSE:', mean_squared_error(Y_test.iloc[:, i], Y_pred[:, i]))
    print(' R2:', r2_score(Y_test.iloc[:, i], Y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'id': [station_id], 'year': [year_input]})
input_data_encoded = pd.get_dummies(input_data, columns=['id'], )
missing_columns = set(X_encoded.columns) - set(input_data_encoded.columns)
for col in missing_columns:
    input_data_encoded[col] = 0
input_data_encoded = input_data_encoded[X_encoded.columns]

predicated_pollutants = model.predict(input_data_encoded)[0]

print(f"\nPredicted pollutants levels for station '{station_id}' in  {year_input}:")
for p, val in zip(pollutants, predicated_pollutants):
  print(f" {p}: {val:.2f}")
```

```python
import joblib

joblib.dump(model, 'pollution_model.pkl')
joblib.dump(X_encoded.columns.tolist(),"model_columns.pkl")
print('Model and cols structure are saved!')
```
